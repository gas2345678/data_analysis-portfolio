```python
import pandas as pd
import seaborn as sb
import numpy as np
```

We load a default datasets list

```python
sb.get_dataset_names()
```

we use the "diamonds" dataset



```python
diamonds_dt = sb.load_dataset("diamonds")
```

```python
diamonds_dt
```

A scatterplot based on price related to depth of diamonds

```python
sb.scatterplot(data=diamonds_dt,x="depth",y="price",hue="cut")
```

a lineplot which represents the relationship between carat of diamonds and their price

```python
sb.lineplot(data=diamonds_dt,x="carat",y="price",dashes=True)
```

we analyse if the price depends on the cut of diamonds and separate them by clarity in a barplot

```python
sb.barplot(data=diamonds_dt,x="cut",y="price",hue="clarity")
```
