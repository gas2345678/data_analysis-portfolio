```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv(r"C:\Users\CR\Downloads\mobile_price.csv")
```

```python
df= pd.DataFrame(df)
```

```python
df
```

```python
sb.lineplot(data=df,x="talk_time",y="battery_power")
plt.xlabel("talk time")
plt.ylabel("power of battery")
plt.title("the effects of talk time in mobile's battery")
```

```python
sb.regplot(data=df,x="ram",y="battery_power")
plt.title("ram influence on battery power")
```
