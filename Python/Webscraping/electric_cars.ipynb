```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests as re
```

```python
url = "https://en.wikipedia.org/wiki/Electric_car"
site = re.get(url)
```

```python
soup = BeautifulSoup(site.text,"html.parser")
```

```python
soup_pretty = soup.prettify()
soup_pretty
```

```python
header_3 = soup.find_all("h3")
header_3
for header in header_3:
    print("-Header",header.text)
```

```python
#We want to find all the tables from the page
table = soup.find_all("table")
#We take each row of the table and extract the text
for element in table:
    table_row = element.find("tr")
    
    table_text = table_row.get_text()
    #We print the text
    print(table_text)
```
