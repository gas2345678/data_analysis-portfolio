```python
#Import the required libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Import the dataset
df = pd.read_csv(r"C:\Users\PC\Desktop\data_analysis-portfolio\datasets\london_houses.csv")
```

```python
#We clean our dataset 
df = df.dropna()
```

```python
#Statistical overview of the data
df.describe()
```

```python
#Dataset shape
df.shape
```

```python
#Retrieve all the apartements 
apartments = df.loc[df["Property Type"] == "Apartment"]
apartments
```

```python
#All apartments with an area less than 100 square meters
small_apartments = df.loc[(df["Property Type"] == "Apartment") & (df["Square Meters"] < 100)]
small_apartments.head(10)
```

We want to figure out if there is more than houses than apartments by comparing both quantities

```python
#Number of properties for each type
houses_apartments = df.groupby("Property Type")["Property Type"].value_counts()
print(houses_apartments)
```

```python
#Visualize the previous results
properties_char = sb.countplot(data=df,x="Property Type",hue="Property Type")
properties_char.set_ylabel("Total")
```

There are certain aspects which can determine the price of the property.

We want to see the correlation between the final price and the area of each property to discover if this is an important factor to consider at the time of buying a property in london.

```python
#Visualization of the relationship between the properties's area and their price
price_scatter = plt.scatter(data=df,x="Square Meters",y="Price (£)",color="red",plotnonfinite=True)
plt.xlabel("Square Meters")
plt.ylabel("Price(k)")
plt.title("Relationship between properties's area and their price")
```

We can observe the bigger the property,the more expensive.

if we want to figure out if the builiding status is also an important factor, we perform the following operations.

```python
sb.set_style("ticks")
grid = sb.FacetGrid(df,col="Property Type",hue="Property Type")
grid.map(sb.stripplot,"Building Status","Price (£)")
```

This scatterplot shows the relationship between the building status and their price and a segmentation based on the type of property.
