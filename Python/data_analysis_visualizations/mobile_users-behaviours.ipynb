The objective of this project is to analyze mobile use of all around the world, focusing on certain parameters like the time spended on the screen, age of participants or the number
of apps installed in their devices.

To achieve these results, we use Pandas and Seaborn libraries to make an EDA based on the data provided.

```python
#Import the libraries
import pandas as pd 
import seaborn as sb 
```

```python
#We search for our dataset
df = pd.read_csv(r"C:\Users\CR\Desktop\datasets_used\data_visualizations\user_behavior_dataset.csv")
```

If we  want to know the total data usage consume by users, we have to get the total data usage for each operative system.This could help us to know which is the most used operative system
in the actual world.


We find out the device with the highest data consume is iOS

```python
#¿Which operating system is the most mobile MB consumer?
total_data_usage = df.groupby("Operating System")["Data Usage (MB/day)"].sum()
total_data_usage
```

```python
#¿Which is the most used Operating System?
most_used_operating_system = df["Operating System"].mode()
most_used_operating_system


```

Now we perfom a countplot to discover the most used OS.

```python

os_users_countplot = sb.countplot(data=df,x="Operating System",hue="Operating System")
os_users_countplot.set_title("Operating Systems users")
os_users_countplot.set_xlabel("mobile operating system")
os_users_countplot.set_ylabel("users")
```

We see although most users use an Android device, this doesn`t mean users is the most data consumer.

To keep going with our EDA, we will group the total time on screen spended by users and group them by their age.
Next, we will make a lineplot to see if there is a relationship between age of participants and hours spended on devices.

```python
#We see if users s' age have an influence on total time on screen
app_usage_time = df["Screen On Time (hours/day)"]
age_usage_time = df.groupby("Age")["Screen On Time (hours/day)"].sum()
#There is not a big influence between these two variables
line_plot_age = sb.lineplot(age_usage_time)
line_plot_age.set_xlabel("users's age")
line_plot_age.set_ylabel("Total Screen Time(hours/day)")
```

It seems there is no direct relationship between these two variables

We can see the maximum time on screen spended was seen in 35 years old users
We can say 35 years old users consume more hours than others users.
