```python
import pandas as pd 
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"C:\Users\CR\Downloads\Boxing.csv")
df
```

# We rename some columns

```python
df.rename(columns={"discipline_name": "discipline","event_name": "event","participant_name": "participant"})
df.replace({"M": "Male", "W": "Female"})
```

# We drop a unnecesary column.In this case is "participant_type"

```python
df.drop(columns = ["participant_type"])
```

# We want to know the total amount of points and group them by gender


```python
sb.countplot(data=df,x="result",hue="gender")


plt.title("results of matches based in points")
plt.legend(labels = ["Male","Female"])
```
