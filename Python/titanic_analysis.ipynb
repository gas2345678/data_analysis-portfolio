# comparison of prices of fares payed by both genders

```python
sb.barplot(data=titanic,x="sex",y="fare",hue="sex")
```

# Percent of people who survived to the accident

```python
sb.countplot(data=titanic,x="alive",hue="sex",stat="percent")
plt.title("titanic survivors")
plt.ylabel("%")
```

# Price of fare to pay based on passengers's class and their embark town

```python
grid = sb.FacetGrid(data=titanic,row="class",col="embark_town")
grid.map_dataframe(sb.histplot,x="fare")
```
